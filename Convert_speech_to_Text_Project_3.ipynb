```python
import librosa
import torch
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
from IPython.display import display, Audio

# Path to the uploaded FLAC file
file_name = '/2902-9008-0000.flac'  # Update path if needed

# Load the processor and model
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")

# Load audio with librosa
try:
    input_audio, _ = librosa.load(file_name, sr=16000)
    print(f"Audio loaded successfully: {file_name}")

    # Process the audio file
    input_values = processor(input_audio, return_tensors="pt", padding=True).input_values

    # Get logits from the model
    logits = model(input_values).logits

    # Get predicted ids from the logits
    predicted_ids = torch.argmax(logits, dim=-1)

    # Decode the predicted ids to get the transcription
    transcription = processor.batch_decode(predicted_ids)[0]

    # Output the transcription
    print("Transcription: ", transcription)

except FileNotFoundError:
    print(f"Error: File '{file_name}' not found. Please upload the file or specify the correct path.")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
